```python
import requests
import pandas as pd

#興櫃

def take_data(type):
    res = requests.get("https://isin.twse.com.tw/isin/C_public.jsp?strMode="+str(type))
    df = pd.read_html(res.text)[0]
    # 設定column名稱
    df.columns = df.iloc[0]
    # 刪除第一行
    df = df.iloc[2:]
    # 先移除row，再移除column，超過三個NaN則移除
    df = df.dropna(thresh=3, axis=0).dropna(thresh=3, axis=1)
    df=df[df['CFICode'] == "ESVUFR"]
    df["有價證券代號及名稱"]=df["有價證券代號及名稱"].apply(lambda x:x.replace(" ","　"))
    
    df["代號"]=df["有價證券代號及名稱"].apply(lambda x:x.split("　")[0])
    df["公司名稱"]=df["有價證券代號及名稱"].apply(lambda x:x.split("　")[1])
    df = df.set_index('有價證券代號及名稱')
    return df
```

```python
def output(df,output_name):
    df.to_csv(output_name,encoding="utf_8_sig")
```

```python
#2 上市  https://isin.twse.com.tw/isin/C_public.jsp?strMode=2
df_1=take_data(2)
#2 上櫃 https://isin.twse.com.tw/isin/C_public.jsp?strMode=4
df_2=take_data(4)
#5 興櫃 https://isin.twse.com.tw/isin/C_public.jsp?strMode=5
df_5=take_data(5)
```

```python
output(df_1,"上市.csv")
output(df_2,"上櫃.csv")
output(df_5,"興櫃.csv")
```
